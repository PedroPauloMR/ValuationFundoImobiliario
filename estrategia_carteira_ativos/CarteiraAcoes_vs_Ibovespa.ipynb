```python
import yfinance as yf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tickers = "TAEE4.SA ITSA4.SA TAEE3.SA EGIE3.SA ENBR3.SA WEGE3.SA HYPE3.SA SQIA3.SA ITSA3.SA"
carteira = yf.download(tickers, period = '19mo')['Adj Close']
ibov = yf.download('^BVSP', period = '19mo')['Adj Close']
bova = yf.download('BOVA11.SA', period = '19mo')['Adj Close']
small = yf.download('SMAL11.SA', period = '19mo')['Adj Close']
```

```python
carteira.dropna(inplace = True)
carteira
```

```python
ibov.dropna(inplace = True)
ibov
```

```python
bova.dropna(inplace = True)
bova
```

```python
small.dropna(inplace = True)
small
```

```python
sns.set()
carteira.plot(figsize=(18,8))
```

```python
carteira_normalizada = (carteira / carteira.iloc[0])*10000
carteira_normalizada.dropna(inplace = True)

carteira_normalizada.plot(figsize = (17,8))
```

```python
carteira_normalizada['saldo'] = carteira_normalizada.sum(axis = 1)
carteira_normalizada
```

```python
ibov_normalizado = (ibov / ibov.iloc[0]) * 90000
ibov_normalizado
```

```python
bova_normalizado = (bova / bova.iloc[0]) * 90000
bova_normalizado
```

```python
small_normalizado = (small / small.iloc[0]) * 90000
small_normalizado
```

```python
carteira_normalizada['saldo'].plot(figsize=(18,8), label = 'Minha Carteira')
ibov_normalizado.plot(label = 'IBOV')
bova_normalizado.plot(label = 'BOVA11')
small_normalizado.plot(label = 'SMAL11')
plt.legend();
```
